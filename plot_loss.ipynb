```python

from Models.model import *

import glob

import os

from train_galaxy_resnet import *

from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt
import glob
```

```python
pre_model_path_list = glob.glob('/data/docker_containers/ljm/GalaxyClassification/GalaxyClassification/results/lens/resnet18*/*/logs/*')

# 定义 9 种颜色
color_list = ['red', 'blue', 'green', 'black', 'yellow', 'purple', 'orange', 'pink', 'brown']

pre_model_path_list = sorted(pre_model_path_list,key=lambda x:os.path.getmtime(x))

plt.figure(figsize=(20, 5))

for index,i in enumerate(pre_model_path_list):


    event_acc = EventAccumulator(i)
    event_acc.Reload()
    print(event_acc.Tags())

    plt.subplot(141)

    scalars = event_acc.Scalars('train_acc')  

    steps = [s.step for s in scalars]
    values = [s.value for s in scalars]

    plt.plot(steps, values, label=i.split("/")[-4],linestyle='-',linewidth=1.5,color=color_list[index])
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    # plt.legend()
    plt.grid(True)
    # plt.ylim([0.6,1])


    scalars = event_acc.Scalars('val_acc')  

    steps = [s.step for s in scalars]
    values = [s.value for s in scalars]


    plt.plot(steps, values,linestyle='--',linewidth=1,color=color_list[index])
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    # plt.ylim([0.6,1])
    plt.legend()
    plt.grid(True)

    plt.subplot(142)

    scalars = event_acc.Scalars('train_loss')

    steps = [s.step for s in scalars]
    values = [s.value for s in scalars]

    plt.plot(steps, values, label=i.split("/")[-4],linestyle='-',linewidth=1.5,color=color_list[index])
    plt.xlabel('Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    # plt.legend()

    scalars = event_acc.Scalars('val_loss')

    steps = [s.step for s in scalars]
    values = [s.value for s in scalars]

    plt.plot(steps, values,linestyle='--',linewidth=1,color=color_list[index])
    plt.xlabel('Epoch')
    plt.ylabel('Loss')

    plt.legend()
    plt.grid(True)

    if "vit" in i:
        plt.subplot(143)

        scalars = event_acc.Scalars('lr')

        steps = [s.step for s in scalars]
        values = [s.value for s in scalars]

        plt.plot(steps, values, label=i.split("/")[-4],linestyle='-',linewidth=1.5,color=color_list[index])
        plt.xlabel('Epoch')
        plt.xlabel('Epoch')
        plt.ylabel('lr')

        plt.legend()
        plt.grid(True)

    else:
        plt.subplot(143)

        scalars = event_acc.Scalars('lr')

        steps = [s.step for s in scalars]
        values = [s.value for s in scalars]

        plt.plot(steps, values, label=i.split("/")[-4],linestyle='-',linewidth=1.5,color=color_list[index])
        plt.xlabel('Epoch')
        plt.xlabel('Epoch')
        plt.ylabel('lr')

        plt.legend()
        plt.grid(True)

plt.show()

```
